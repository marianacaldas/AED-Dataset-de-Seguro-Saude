# Introdução

Este conjunto de dados de seguro saúde contém informações detalhadas sobre segurados, como idade, sexo, índice de massa corporal (IMC), número de filhos, fumantes e região de residência. Além disso, o dataset inclui o valor cobrado pela apólice de seguro para cada indivíduo.

A análise deste conjunto de dados visa identificar os principais fatores que influenciam o custo do seguro saúde, com o objetivo de:
- Explorar como características demográficas e comportamentais, como idade, IMC e status de fumante, impactam o valor do seguro.
- Examinar padrões geográficos que possam estar associados a variações no custo.
- Desenvolver modelos preditivos para estimar o custo de novos segurados com base em suas características.
- 
Ao aplicar técnicas de estatística descritiva e visualização de dados, este projeto busca fornecer insights valiosos para otimizar a precificação de seguros e identificar potenciais grupos de risco.

O desvio padrão indica a dispersão dos dados em relação à média, ou seja, quanto as despesas individuais com seguro de saúde variam em torno da média para cada grupo.

- Não fumantes:
    Mulheres (female): O desvio padrão é 6060.78, o que sugere que os custos variam significativamente, mas menos que nos fumantes.
    Homens (male): O desvio padrão é 5908.11, um pouco menor que o das mulheres, indicando uma dispersão ligeiramente menor nos custos dos homens não fumantes.
  
- Fumantes:
    Mulheres (female): O desvio padrão é 11907.54, mostrando uma dispersão maior, sugerindo que há uma variação muito maior no custo dos seguros de saúde entre mulheres fumantes.
    Homens (male): O desvio padrão é 11202.67, também bastante alto, mas ainda menor que o das mulheres fumantes, indicando que a variação entre os custos dos homens fumantes é muito grande, mas menor do que a variação entre as mulheres fumantes.

Em resumo, os fumantes têm um desvio padrão muito maior, indicando que os custos de seguro de saúde variam mais para os fumantes do que para os não fumantes. Além disso, a variação entre as mulheres fumantes é maior que entre os homens fumantes.

```python
smoke_counts = df['smoker'].value_counts()

plt.figure(figsize=(8, 6))

plt.bar(
    smoke_counts.index,
    smoke_counts.values,
    color=['steelblue', 'lightcoral']
)

plt.xlabel('Smoker', fontsize=8)
plt.ylabel('Count', fontsize=8)
plt.title('Distribuição de Fumantes', fontsize=14)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

- Não fumantes: A maior parte da amostra é composta por não fumantes, com uma contagem superior a 1000 indivíduos.
- Fumantes: Aproximadamente 200 a 300 indivíduos são fumantes, o que é uma parcela bem menor em relação aos não fumantes.

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(df['age'], df['charges'], c="red")

ax.set_title('Gráfico de Dispersão Idade vs Cobrança', fontsize=16)
ax.set_xlabel('Idade', fontsize=12)
ax.set_ylabel('Cobranças', fontsize=12)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

O gráfico de dispersão correlaciona a idade com os valores cobrados. Observa-se que os valores de cobrança parecem aumentar com a idade de forma ascendente , com padrões de dispersão em faixas específicas.

```python
fig, axes = plt.subplots(
    1,
    2,
    figsize=(12, 6)
)

axes[0].scatter(df['bmi'], df['charges'])
axes[0].set_xlabel('BMI')
axes[0].set_ylabel('Charges')
axes[0].set_title('Charges vs IMC')

axes[1].scatter(df['age'], df['charges'])
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Charges')
axes[1].set_title('Charges vs Age')

plt.tight_layout()

plt.show()
```

No gráfico de dispersão Charges vs IMC: 
    Existe uma leve tendência positiva: à medida que o IMC aumenta, há um aumento geral nas cobranças.
    A maioria dos pontos está concentrada em valores de IMC entre 20 e 35.
    Há uma notável dispersão nas cobranças, especialmente para IMCs acima de 30.
    Existem alguns outliers com cobranças muito altas (acima de 40.000) para IMCs variados acima de 30.

No gráfico de dispersão Chages vs Age:
    Há uma tendência positiva mais clara: as cobranças tendem a aumentar com a idade.    
A dispersão das cobranças aumenta significativamente com a idade, indicando maior variabilidade para segurados mais velhos.    
Existe uma concentração de pontos formando uma "banda" que se move diagonalmente para cima, sugerindo uma relação mais forte entre idade e cobranças comparada ao IMC    .
Há outliers notáveis, com algumas cobranças muito altas em várias faixas etár, mas principalmente acima de 30 anosi- Comparação dos dois gráficos:
    A idade parece ser um fator mais determinante para as cobranças de seguro de saúde do que o IMC.   
Ambos os fatores contribuem para o aumento das cobranças, mas de maneiras diferentes.        
A presença de outliers e a dispersão variável sugerem que outros fatores não mostrados nesses gráficos (como condições de saúde preexistentes, hábitos de vida, etc.) também podem ter impactos significativos nas cobrançaas.ear.

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

sns.countplot(data=df, x='smoker', hue='smoker', ax=ax[0, 0], palette='Blues', legend=False)
ax[0, 0].set_title('Fumantes')
sns.countplot(data=df, x='children', hue='children', ax=ax[0, 1], palette='Greens', legend=False)
ax[0, 1].set_title('Filhos')
sns.countplot(data=df, x='region', hue='region', ax=ax[1, 0], palette='Purples', legend=False)
ax[1, 0].set_title('Região')
sns.countplot(data=df, x='sex', hue='sex', ax=ax[1, 1], palette='Oranges', legend=False)
ax[1, 1].set_title('Sexo')

plt.tight_layout()
plt.show()
```

Os gráficos acima apresentam uma análise visual de 4 variáveis categóricas:

- Fumantes:
    A maioria das pessoas analisadas não são fumantes(mais de 1000), enquanto uma pequena parcela (menos de 300) se identificam como fumantes;

- Filhos:
    A maior parte dos indivíduos não tem filhos (cerca de 550), seguido por pessoas com 1 ou 2 filhos;
    Conforme o número de filhos aumenta, o número de indivíduos cai drasticamente.

- Região:
    As quatro regiões apresentam distribuições bastante similares, com uma leve maior concentração de pessoas na região southeast;
    Não há uma diferença muito significativa entre as regiões, sugerindo uma distribuição quase uniforma.

- Sexo:
    O número de homens e mulheres na amostra está praticamente equilibrado, com cerca de 650 indivíduos de casa sexo.

# Conclusão

- Os custos de seguro saúde variam mais para os fumantes do que para os não fumantes. Além disso, a variação entre mulheres fumantes é maior que entre os homens fumantes;
- A idade parece ser um fator mais determinante para as cobranças de seguro saúde do que o IMC;
- Tanto a idade como o IMC contribuem para o aumento das cobranças, mas de maneiras diferentes;
- A presença de outliers e a dispersão variável sugerem que outros fatores não mostrados nesses gráficos (como condição de saúde pré-existente, hábitos de vida, etc.) também podem ter impactos significativos nas cobranças;
